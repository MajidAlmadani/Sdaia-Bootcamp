#**Python Code Implementation of t-SNE on MNIST Dataset**


#Step 1 :
##Importing Necessary Modules.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

# Step 2 :
## load the MNIST_test dataset into pandas dataframe

```python
df = pd.read_csv('dataset/mnist_test.csv')
print(df.head(5))
```

```python
d = df.drop("label", axis=1)
```

# Step 3  :
## Standardize the data

```python
standardized_data = StandardScaler().fit_transform(d)
print(standardized_data.shape)
```

# Step 4 :  Reduce the columns data to 2 dimensions




```python
data_1000 = standardized_data[0:1000, :]
```

```python
labels = df['label'] 
labels_1000 = labels[0:1000]
```

# Step 5 :
## define the model

```python
model = TSNE(n_components = 2, random_state = 0)
```

# Step 6 :
## Fit the model

```python
tsne_data = model.fit_transform(data_1000)
tsne_data = np.vstack((tsne_data.T, labels_1000)).T
tsne_df = pd.DataFrame(data = tsne_data,columns =("Dim_1", "Dim_2", "label"))
```

# Step 7 :
##  Plotting the result

```python
import seaborn as sn
import matplotlib.pyplot as plt
sn.scatterplot(data=tsne_df, x='Dim_1', y='Dim_2',hue='label', palette="bright")
plt.show()
```
